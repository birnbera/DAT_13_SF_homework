```python
from __future__ import division
import pandas as pd
import numpy as np
import re
from mpl_toolkits.basemap import pyproj
```

```python
def qc(cell):
    match = re.search(r'^\[\[[-\d,\.\[\]]+\]\]$', cell)
    if match:
        return np.asarray(eval(cell))
    else:
        return np.nan
```

```python
def get_lons_lats(df):
    lons = df.POLYLINE.apply(lambda x: x[:,0].ravel())
    lats = df.POLYLINE.apply(lambda x: x[:,1].ravel())
    ylon = df.Y.apply(lambda x: x[0])
    ylat = df.Y.apply(lambda x: x[1])
    df.drop('POLYLINE', axis='columns', inplace=True)
    df.drop('Y', axis='columns', inplace=True)
    df['LONS'] = lons
    df['LATS'] = lats
    df['Y_LON'] = ylon
    df['Y_LAT'] = ylat
    return df
```

```python
def trunc_chunk(cell, min_feats=20, trunc_pct=50):
    l = len(cell)
    z = np.zeros((min_feats))
    if l <= min_feats:
        z[min_feats-l:] = cell[:]
        z[:min_feats-l] = cell[0]
    elif l <= 2*min_feats:
        z[:] = cell[:min_feats]
    else:
        tl = (l*trunc_pct)//100
        z[:] = cell[tl-min_feats:tl]
    return z
```

```python
def trunc_for_pred(cell, min_feats=20):
    # Different than `trunc_chunk` since you really want the more recent 20 points, not a truncated polyline
    l = len(cell)
    z = np.zeros((min_feats))
    if l <= min_feats:
        z[min_feats-l:] = cell[:]
        z[:min_feats-l] = cell[0]
    else:
        z[:] = cell[-min_feats:]
    return z
```

```python
chunk.LONS = chunk.LONS.apply(trunc_chunk)
```

```python
chunk.LATS = chunk.LATS.apply(trunc_chunk)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import KFold, train_test_split
from sklearn.metrics import mean_squared_error

def series_to_feats(df, label):
    s = df[label]
    s = np.vstack([ss for ss in s.as_matrix()])
    new_df = {'{}_{}'.format(label,k):v for k,v in zip(range(1,21),[s[:,i] for i in range(s.shape[1])])}
    new_df = pd.DataFrame(new_df, index=df.index)
    retdf = pd.concat([df, new_df], axis=1)
    retdf.drop(label, axis='columns', inplace=True)
    return retdf
```

```python
def kaggle_haversine(coord1, coord2, r = 6371.0):
    lat1, lon1 = coord1
    lat2, lon2 = coord2
    lat = abs(lat1-lat2)*np.pi/180
    lon = abs(lon1-lon2)*np.pi/180
    lat1 = lat1*np.pi/180
    lat2 = lat2*np.pi/180
    a = np.sin(lat/2)*np.sin(lat/2)+np.cos(lat1)*np.cos(lat2)*np.sin(lon/2)*np.sin(lon/2)
    d = 2*np.arctan2(np.sqrt(a),np.sqrt(1-a))
    d = r*d
    return(d)
```

```python
Y_lons = chunk.Y_LON
Y_lats = chunk.Y_LAT
lat_dat = chunk.drop(['TAXI_ID', 'LONS', 'Y_LON'], axis = 'columns')
lon_dat = chunk.drop(['TAXI_ID', 'LATS', 'Y_LAT'], axis = 'columns')
lat_dat = series_to_feats(lat_dat, 'LATS')
lon_dat = series_to_feats(lon_dat, 'LONS')
```

```python
lat_dat = lat_dat.drop('Y_LAT', axis='columns')
lon_dat = lon_dat.drop('Y_LON', axis='columns')
```

```python
lon_dat_train, lon_dat_test, lon_Y_train, lon_Y_test = train_test_split(lon_dat, Y_lons, test_size=0.2, random_state=1)
lat_dat_train, lat_dat_test, lat_Y_train, lat_Y_test = train_test_split(lat_dat, Y_lats, test_size=0.2, random_state=1)
```

```python
lon_reg = RandomForestRegressor(5000, n_jobs=-1, max_depth=20, random_state=1)
lon_reg.fit(lon_dat_train, lon_Y_train)
lon_preds = lon_reg.predict(lon_dat_test)
mse = mean_squared_error(lon_Y_test, lon_preds)
print 'Longitude MSE: {}'.format(mse)

lat_reg = RandomForestRegressor(5000, n_jobs=-1, max_depth=20, random_state=1)
lat_reg.fit(lat_dat_train, lat_Y_train)
lat_preds = lat_reg.predict(lat_dat_test)
mse = mean_squared_error(lat_Y_test, lat_preds)
print 'Latitude MSE: {}'.format(mse)

hdists = []
true_coords = zip(lat_Y_test, lon_Y_test)
pred_coords = zip(lat_preds, lon_preds)
for c1,c2 in zip(true_coords, pred_coords):
    dist = kaggle_haversine(c1,c2)
    hdists.append(dist)
print 'Mean haversine distance: {:.3f}'.format(np.mean(hdists))
```

```python
def kaggle_predict(lon_reg, lat_reg, test_data_path):
    '''Predict destination of the test set based on a pair of 
    trained regressors from sklearn implenting a predict method'''
    td = pd.read_csv(test_data_path)
    td.POLYLINE = td.POLYLINE.apply(qc)
    td.dropna(axis='rows', subset=['POLYLINE'], inplace=True)
    trip_ids = td.TRIP_ID.values
    td.drop(['TRIP_ID', 'MISSING_DATA', 'ORIGIN_CALL','ORIGIN_STAND'], axis = 'columns', inplace=True)
    td = pd.get_dummies(td, columns=['CALL_TYPE', 'DAY_TYPE'])
    td['Y'] = td.POLYLINE.apply(lambda x: x[-1])
    td = get_lons_lats(td)
    td.LONS = td.LONS.apply(trunc_for_pred)
    td.LATS = td.LATS.apply(trunc_for_pred)
    Y_lons = td.Y_LON
    Y_lats = td.Y_LAT
    lat_dat = td.drop(['TAXI_ID', 'LONS', 'Y_LON'], axis = 'columns')
    lon_dat = td.drop(['TAXI_ID', 'LATS', 'Y_LAT'], axis = 'columns')
    lat_dat = series_to_feats(lat_dat, 'LATS')
    lon_dat = series_to_feats(lon_dat, 'LONS')
    lat_dat = lat_dat.drop('Y_LAT', axis='columns')
    lon_dat = lon_dat.drop('Y_LON', axis='columns')
    lon_preds = lon_reg.predict(lon_dat)
    lat_preds = lat_reg.predict(lat_dat)
    path = '/Users/Bronco/Classes/General_assembly/DAT_SF_13_homework/Final_project/kaggle/taxi_trajectory_prediction/'
    with open(path+'submission.csv', 'w') as sub:
        sub.write('"TRIP_ID","LATITUDE","LONGITUDE"\n')
        for c in zip(trip_ids,lat_preds,lon_preds):
            write = '"{}",{:.6f},{:.6f}\n'.format(c[0],c[1],c[2])
            sub.write(write)
```
